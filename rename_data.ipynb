```python
import os
import re

paths = [
    "IDP Group A/test1/operator1",
    "IDP Group A/test1/operator2",
    "IDP Group A/test1/operator3",
    "IDP Group A/test2/operator1",
    "IDP Group A/test2/operator2",
    "IDP Group A/test2/operator3",
]
for path in paths:
    for name in os.listdir(path):
        old_path = os.path.join(path, name)

        if os.path.isdir(old_path) and re.fullmatch(r"d\d+", name):
            new_name = "s" + name[1:]
            new_path = os.path.join(path, new_name)

            os.rename(old_path, new_path)
            print(f"{old_path} -> {new_path}")
```
